

The arithmetic sequence, 1487, 4817, 8147, in which each of the terms increases by 3330, is unusual in two ways: (i) each of the three terms are prime, and, (ii) each of the 4-digit numbers are permutations of one another.

There are no arithmetic sequences made up of three 1-, 2-, or 3-digit primes, exhibiting this property, but there is one other 4-digit increasing sequence.

What 12-digit number do you form by concatenating the three terms in this sequence?


```python
## import math as math
import numpy as np
import itertools as iter

def isPrime(number):  
    if number<=1 or number%2==0:  
        return 0  
    check=3
    maxneeded=number  
    while check<maxneeded+1:  
        maxneeded=number/check  
        if number%check==0:  
            return False
        check+=2  
    return True

x=[]
for i in range(1000,10000):
    if isPrime(i):
        x.append(i)

print(len(x))


def numlist(z):
    xlist=[]
    xlist.append(int(z/1000))
    xlist.append(int(z/100%10))
    xlist.append(int(z/10%10))
    xlist.append(z%10)
    
    sorted=np.argsort(xlist)
    return(xlist)

z=3215
print(numlist(z))
print(isPrime(z))
#x=[1487]
allprimes=[]
for z in x:
    perms = [ perms for perms in iter.permutations(numlist(z))]
    permlist=[]
    for perm in perms:
        permstr=""
        for index in range(4):
            permstr+=str(list(perm)[index])
        permint=int(permstr)
        if permint > 999: 
            allprimes.append(permint)
            #print(permint)
            if isPrime(permint) and permint not in permlist:
                permlist.append(permint)
        permlist.sort()
    if len(permlist)>=3: 
        #print(z,permlist)
        diffs=[]
        for index in range(1,len(permlist)):
            diffs.append(permlist[index]-permlist[index-1])
        diffcombos=[]
        for index in range(1,8):
            difflist = [sum(x1) for x1 in iter.combinations(diffs,index)]
            for tmpdiff in difflist:
                diffcombos.append(tmpdiff)
        #print(diffcombos)
        for diff in diffcombos:
            #print("     ",diff)
            for index in range(len(permlist)):
                if (permlist[index]-diff) in permlist and (permlist[index]-diff*2) in permlist:
                    print(permlist[index]-diff*2,permlist[index]-diff,permlist[index],"-",diff)
        
```
